### **Reading Data using Panda**

```python
#Import Libraries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

# Read in the data
df = pd.read_csv('/content/movies.csv')
```

```python
# View Data
df.head()
```

### **Looking For Missing Data**

```python
df = df.dropna()
```

```python
# Checking for missing data

for col in df.columns:
  pct_missing = np.mean(df[col].isnull())
  print('{} - {}%'.format(col,pct_missing))
```

```python
# Check Data Type sof Columns

df.dtypes
```

### **Data Cleaning**

```python
# casting data types of columns

df['budget'] = df['budget'].astype('int64')

df['gross'] = df['gross'].astype('int64')
```

**Note:**  Won't cast if there are null values in the dataset. Gives error.
Remove null before casting

`df = df.dropna()`

```python
df.head()
```

Decimals from gross and budget is removed.

Budget and gross are converted to integer.

To get the proper year because of different Date Format

```python
# Creating Year Column

df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)
```

**OR**

```python
df['yearcorrect'] = df['released'].astype(str).str.split(', ').str[-1].astype(str).str[:4]
```

```python
df.head()
```

**Note:** There is a wrong year column in dataset

**Sorting by Gross Column**

```python
# gross column is sorted in descending order

df = df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# to get all the data

pd.set_option('display.max_rows', None)
```

**Removing Duplicates**

```python
# Viewing the duplicates first
df['company'].sort_values(ascending=False)
```

Gives the number of duplicate counts ⬇

```python
# Dropping Duplicates

df['company'].drop_duplicates().sort_values(ascending=False).head()
```

**Correlation**

**Different types of Correlation**

1. Pearson (by default)
2. Kendall,
3. Spearman

```python
df.corr(method="pearson")
```

```python
df.corr(method="kendall")
```

```python
df.corr(method="spearman")
```

      High Correlation between budget and gross

```python
correlation_matrox = df.corr(method="pearson")

sns.heatmap(correlation_matrox, annot=True)

plt.title('Correlation Matrix for Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```

```python
# Viewing Company

df.head()
```

    Having Numeric Representation of the company





```python
df_numerized = df

for col_name in df_numerized.columns:
  if (df_numerized[col_name].dtype == 'object'):
    #renaming objects
    df_numerized[col_name] = df_numerized[col_name].astype('category') 
    df_numerized[col_name] = df_numerized[col_name].cat.codes

df_numerized.head()
```

```python
correlation_matrox = df_numerized.corr(method="pearson")

sns.heatmap(correlation_matrox, annot=True)

plt.title('Correlation Matrix for Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```

```python
df_numerized.corr()
```

**Unstacking**

```python
correlation_mat = df_numerized.corr()

corr_pairs = correlation_mat.unstack()
# corr_pairs
corr_pairs.head()
```

Sorting above correlation pairs

```python
sorted_pairs = corr_pairs.sort_values()
# sorted_pairs
sorted_pairs.head()
```

```python
high_corr = sorted_pairs[(sorted_pairs)>0.5]
high_corr
```

> Votes and budget have the highest correlation to gross earnings

> Company has low correlation
